```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from palettable.cartocolors.sequential import Magenta_7, PurpOr_7
from palettable.colorbrewer.diverging import BrBG_11
from palettable.colorbrewer.sequential import RdPu_9
from palettable.scientific.sequential import Bamako_16_r, Davos_16_r, Tokyo_18_r

Magenta_7.mpl_colormap
```

```python
#load data
df = pd.read_csv("US_births_1994-2014.csv")

data = df.groupby(["month", "date_of_month"])["births"].mean().unstack()
data.head()
```

## Resources

Matplotlib: https://matplotlib.org/stable/tutorials/colors/colormaps.html

Palettable: https://jiffyclub.github.io/palettable/

```python
weekdays= ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
month_labels =  [ 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

## plot
fig, ax = plt.subplots(figsize=(10,4), layout='tight')
ax = sns.heatmap(data, cmap=Davos_16_r.mpl_colormap, robust=True, cbar=False)

#========= Axis
#set weekday labels 
ax.set_yticklabels(month_labels, rotation = 0, fontsize = 10.5, va='center')
ax.tick_params(axis='both', length = 0)
ax.set_ylabel("")
ax.set_xlabel("")

plt.show()
```
