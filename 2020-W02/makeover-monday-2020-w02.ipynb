```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle #for waffle chart
```

```python
# create dataframes
europe = pd.DataFrame({'colour': ['Yes', 'No'],
       'number': [27, 73]
     })

china = pd.DataFrame({'colour': ['Yes', 'No'],
       'number': [3, 97]
     })

brazil = pd.DataFrame({'colour': ['Yes', 'No'],
       'number': [2, 98]
     })
```

```python
# Create waffle chart using pywafffle library
cols = {"yes": "#9e2a2b", "no": "#c3c3c3", "title": "#540b0e", "label": "#012a4a"}

# =========
# plot charts
fig = plt.figure(
    FigureClass=Waffle,
    plots={
        131: {
            'values': europe.number,
            'title': {
                'label': 'Europe',
                'loc': 'center',
                'y':1.03,
                'fontfamily':"arial",
                'fontsize':13,
                'fontweight': 'bold',
                'color': "#111111"
            }
        },
        132: {
            'values': china.number,
            'title': {
                'label': 'China',
                'loc': 'center',
                'y':1.03,
                'fontfamily':"arial",
                'fontsize':13,
                'fontweight': 'bold',
                'color': "#111111"
            }
        },
        133: {
            'values': brazil.number,
            'title': {
                'label': 'Brazil',
                'loc': 'center',
                'y':1.03,
                'fontfamily':"arial",
                'fontsize':13,
                'fontweight': 'bold',
                'color': "#111111"
            }
        },
    },
    rows=10,
    vertical=True,
    colors=(cols["yes"], cols["no"]),  
    figsize=(9, 5) 
)


## Add % labels
plt.figtext(0.13,0.52,'26.9%',
         fontsize = 30, weight = 'bold', fontfamily='Arial', color=cols["label"])
plt.figtext(0.46,0.52,'3.3%',
         fontsize = 30, weight = 'bold', fontfamily='Arial', color=cols["label"])
plt.figtext(0.78,0.52,'2.2%',
         fontsize = 30, weight = 'bold', fontfamily='Arial', color=cols["label"])

## Add lbs labels
plt.figtext(0.13,0.45,'(323m lbs)',
         fontsize = 18, weight = 'normal', fontfamily='Arial', color=cols["label"])
plt.figtext(0.455,0.45,'(40m lbs)',
         fontsize = 18, weight = 'normal', fontfamily='Arial', color=cols["label"])
plt.figtext(0.775,0.45,'(26m lbs)',
         fontsize = 18, weight = 'normal', fontfamily='Arial', color=cols["label"])

# =====
# Add title background colour
fig.set_facecolor("#f7f6f4")
plt.figtext(0.02,0.99,'USA lags behind other agricultural nations in banning harmful pesticides',
         fontsize = 18, weight = 'bold', fontfamily='Arial', color=cols["label"])
plt.figtext(0.02,0.93, "The amount of pesticides used in the US in 2016 that were banned in other countries", 
            fontsize=13, ha='left',  color="#111111", fontfamily="Arial")

#footer
plt.figtext(0.02,0.15, "Makeover Monday 2020-W02 | Data Source: BMC Environmental Health", fontsize=10.5, ha='left', color="#111111", fontfamily="Arial")
plt.figtext(0.81,0.15, "Design by: Lisa Hornung            ", fontsize=10.5, ha='left', color="#111111", fontfamily="Arial")


#save outfile
plt.savefig("MakeoverMonday2020W02_use-of-pesticides-USA.png", bbox_inches="tight",pad_inches=0.2)

plt.show()
```
